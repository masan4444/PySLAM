```python
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
def readCosts(filepath):
    with open(filepath) as f:
        reader = csv.reader(f)
        estCount = 0
        optCount = 0
        costs = []
        _costs = []
        __costs = []
        for row in reader:
            if row[0] == "!":
                # optCount = int(row[1])
                __costs.append(float(row[2]))
            else:
                # estCount = int(row[0])
                # optCount = 0
                _costs.append(__costs)
                __costs = []
                if row[0] == "0":
                    costs.append(_costs)
                    _costs = []
    return costs
```

```python
def readCostsAndDiff(filepath):
    with open(filepath) as f:
        reader = csv.reader(f)
        estCount = 0
        optCount = 0
        costs = []
        diffs = []
        _costs = []
        _diffs = []
        __costs = []
        __diffs = []
        for row in reader:
            if row[0] == "!":
                # optCount = int(row[1])
                __costs.append(float(row[2]))
                __diffs.append([float(row[5]), float(row[6]), float(row[7]), float(row[8]), float(row[9]), float(row[10])])
            else:
                # estCount = int(row[0])
                # optCount = 0
                _costs.append(__costs)
                _diffs.append(__diffs)
                __costs = []
                __diffs = []
                if row[0] == "0":
                    costs.append(_costs)
                    diffs.append(_diffs)
                    _costs = []
                    _diffs = []
    return costs, diffs
```

```python
def getCount(costs):
    counts = []
    _counts = 0
    for _costs in costs:
        for __costs in _costs:
            _counts += len(__costs)
        counts.append(_counts)
        _counts = 0
    return counts
```

```python
def getEstCount(costs):
    counts = []
    for _costs in costs:
        counts.append(len(_costs))
    return counts
```

```python
costs, diffs = readCostsAndDiff('../LittleSLAM/result/1603158922.csv')
counts = getCount(costs)
estCounts = getEstCount(costs)
print(len(counts))
plt.plot(counts, label='accurate multyply')
plt.xlabel('Scan index')
plt.ylabel('Number of times of optimization')
plt.title('Number of times of optimization')
plt.legend()
plt.show()
plt.plot(estCounts)
```
